```python

import argparse
import sys
import datetime



if sys.version_info[0] == 2:
	input = raw_input
    
class Vehicle:
	def __init__(self,regno,types,status,intime,user):
		self.types =  types
		self.regno = regno
		self.status=status
		self.intime =intime
		self.outtime =None
		self.user=user
		self.fee =0.0
        
        
class FourWheeler(Vehicle):

	def __init__(self,regno,types,status,intime,user):
		Vehicle.__init__(self,regno,types,status,intime,user)

	def getType(self):
		return "Four Wheeler"
	def setOuttime(self):
		self.outtime = datetime.datetime.now().replace(microsecond=0)
		c = self.outtime - self.intime
		self.fee = 1 * (c.total_seconds() / 60)
class TwoWheeler(Vehicle):

	def __init__(self,regno,types,status,intime,user):
		Vehicle.__init__(self,regno,types,status,intime,user)

	def getType(self):
		return "Two Wheeler"
	def setOuttime(self):
		self.outtime = datetime.datetime.now().replace(microsecond=0)
		c = self.outtime - self.intime
		self.fee = 1 * (c.total_seconds() / 60)
    
class Large(Vehicle):

	def __init__(self,regno,types,status,intime,user):
		Vehicle.__init__(self,regno,types,status,intime,user)

	def getType(self):
		return "Large Vehicle"
	def setOuttime(self):
		self.outtime = datetime.datetime.now().replace(microsecond=0)
		c = self.outtime - self.intime
		self.fee = 1 * (c.total_seconds() / 60)

class ParkingLot:
	def __init__(self,n):
		self.capacity = 0
		self.slotid = 0
		self.numOfOccupiedSlots = 0
		self.logSlots = []
		self.createParkingLot(n)

	def createParkingLot(self,capacity):
		self.slots = [-1] * capacity
		self.capacity = capacity
		return self.capacity

	def getEmptySlot(self):
		for i in range(len(self.slots)):
			if self.slots[i] == -1:
				return i

	def park(self,regno,types,status,user):
		if self.numOfOccupiedSlots < self.capacity: 
			slotid = self.getEmptySlot()
			if types == "0":
				self.slots[slotid] = TwoWheeler(regno,"Two Wheeler",status,datetime.datetime.now().replace(microsecond=0),user)
			elif types == "1":
				self.slots[slotid] = FourWheeler(regno,"Four Wheeler",status,datetime.datetime.now().replace(microsecond=0),user)
			else:
				self.slots[slotid] = Large(regno,"Large Vehicle",status,datetime.datetime.now().replace(microsecond=0),user)
			self.slotid = self.slotid+1
			self.numOfOccupiedSlots = self.numOfOccupiedSlots + 1
			return slotid+1
		else:
			return -1

	def leave(self,slotid):

		if self.numOfOccupiedSlots > 0 and self.slots[slotid-1] != -1:
			self.slots[slotid-1].setOuttime()
			self.logSlots.append(self.slots[slotid-1])
			self.slots[slotid-1] = -1
			self.numOfOccupiedSlots = self.numOfOccupiedSlots - 1
			return True
		else:
			return False

	def status(self):
		print("User\tSlot No.\tRegistration No.\tType\tstatus.\t\t\tIn Time\t\t\tOutTime\t\t\tFee")
		for i in range(len(self.slots)):
			if self.slots[i] != -1:
				print(str(self.slots[i].user)+ "\t\t"  +str(i+1) + "\t\t" +str(self.slots[i].regno) + "\t\t" + str(self.slots[i].types)+ "\t\t" + str(self.slots[i].status)+ "\t" + str(self.slots[i].intime)+ "\t" + str(self.slots[i].outtime)+ "\t" + str(self.slots[i].fee))
			else:
				continue
                
	def log(self):
		print("User\tSlot No.\tRegistration No.\tType\tstatus.\t\t\tIn Time\t\t\tOutTime\t\t\tFee")
		for i in range(len(self.logSlots)):
			if self.logSlots[i] != -1:
				print(str(self.logSlots[i].user) + "\t\t" +str(i+1) + "\t\t" +str(self.logSlots[i].regno) + "\t\t" + str(self.logSlots[i].types)+ "\t\t" + str(self.logSlots[i].status)+ "\t" + str(self.logSlots[i].intime)+ "\t" + str(self.logSlots[i].outtime)+ "\t" + str(self.logSlots[i].fee))
			else:
				continue

	def getRegNoFromTypes(self,types):

		regnos = []
		for i in self.slots:

			if i == -1:
				continue
			if i.types == types:
				regnos.append(i.regno)
		return regnos
			
	def getSlotNoFromRegNo(self,regno):
		
		for i in range(len(self.slots)):
			if self.slots[i].regno == regno:
				return i+1
			else:
				continue
		return -1
			

	def getSlotNoFromTypes(self,types):
		
		slotnos = []

		for i in range(len(self.slots)):

			if self.slots[i] == -1:
				continue
			if self.slots[i].types == types:
				slotnos.append(str(i+1))
		return slotnos

	def show(self,line):

		if line.startswith('park'):
			regno = line.split(' ')[1]
			types = line.split(' ')[2]
			status = line.split(' ')[3]
			user = line.split(' ')[4]
			res = self.park(regno,types,status,user)
			if res == -1:
				print("Sorry, parking lot is full")
			else:
				print('Allocated slot number: '+str(res))

		elif line.startswith('leave'):
			leave_slotid = int(line.split(' ')[1])
			status = self.leave(leave_slotid)
			if status:
				print('Slot number '+str(leave_slotid)+' is free')

		elif line.startswith('status'):
			self.status()
            
		elif line.startswith('log'):
			self.log()

		elif line.startswith('registration_numbers_for_cars_with_colour'):
			types = line.split(' ')[1]
			regnos = self.getRegNoFromTypes(types)
			print(', '.join(regnos))

		elif line.startswith('slot_numbers_for_cars_with_colour'):
			types = line.split(' ')[1]
			slotnos = self.getSlotNoFromTypes(types)
			print(', '.join(slotnos))

		elif line.startswith('slot_number_for_registration_number'):
			regno = line.split(' ')[1]
			slotno = self.getSlotNoFromRegNo(regno)
			if slotno == -1:
				print("Not found")
			else:
				print(slotno)
		elif line.startswith('exit'):
			exit(0)

def main():

    parkinglot = ParkingLot(5)
    while True:
        line = input("$ ")
        if line =="q":
            break
        parkinglot.show(line)

if __name__ == '__main__':
    main()
```
